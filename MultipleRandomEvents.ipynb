```python
import random
import matplotlib.pyplot as plt

def flip_coin():
    return random.choice(['Heads', 'Tails'])

def roll_die():
    return random.randint(1, 6)

def calculate_probability(num_iterations):
    event_counter = 0  # Counter for the desired event
    probability_list = []  # List to store the calculated probabilities
    true_probability = 1 / 2 * 1 / 6  # True probability of the desired event

    for i in range(1, num_iterations + 1):
        coin_result = flip_coin()
        die_result = roll_die()

        if coin_result == 'Tails' and die_result == 5:
            event_counter += 1

        current_probability = event_counter / i
        probability_list.append(current_probability)

    return probability_list

# Set the number of iterations
num_iterations = 3000

# Calculate the probabilities
probabilities = calculate_probability(num_iterations)

# Plotting the convergence to the true probability
plt.plot(range(1, num_iterations + 1), probabilities)
plt.axhline(y=1/12, color='r', linestyle='--', label='True Probability')  # True probability of getting Tails and 5
plt.xlabel('Number of Iterations')
plt.ylabel('Probability')
plt.title('Convergence to True Probability')
plt.legend()
plt.show()
```
